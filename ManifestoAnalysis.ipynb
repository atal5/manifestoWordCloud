```python
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
raw = parser.from_file('BJP-Election-2019-english.pdf')
text = raw['content']
wordcloud = WordCloud(width=1600, height=800,max_font_size=200, max_words=200,background_color="white").generate(text)
#plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.savefig("bjp_world_cloud.png")
wordcloud.to_file("bjp_world_cloud.png")
plt.show()
```
